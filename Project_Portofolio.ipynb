# Missing Value Handling

```python
data_spotify.isna().sum()
```

dari tabel, dapat dilihat bahwa nilai yang tidak ada (NA value) berada pada label Top_Genre dan Repeat_Song_Rate

## *Fill Missing Value at Top_Genre*

karena nilainya berupa "string", maka kita mengisinya menggunakan modus

```python
data_spotify["Top_Genre"] = data_spotify["Top_Genre"].fillna("Other")
```

```python
data_spotify["Top_Genre"].isna().sum()
```

```python
data_spotify["Top_Genre"].head()
```

Data pada label "Top_Genre" sudah terisi semua, maka kita akan lanjut untuk mengisi NA pada label lainnya

## *Fill Missing Value at Repeat_Song_Rate, Group By Country*

kita mengisi NA pada Repeat_Song_Rate berdasarkan negara, agar dapat terdefinisi lebih jelas, kita juga mengisinya dengan merata-ratakan nilai per negaranya

```python
data_spotify["Repeat_Song_Rate"] = data_spotify.groupby("country")["Repeat_Song_Rate"].transform(lambda x: x.fillna(x.median()))
data_spotify["Repeat_Song_Rate"] = data_spotify["Repeat_Song_Rate"].fillna(data_spotify["Repeat_Song_Rate"].median())
```

```python
data_spotify["Repeat_Song_Rate"].isna().sum()
```

```python
data_spotify["Repeat_Song_Rate"].head()
```

Data pada label "Repeat_Song_Rate" sudah terisi semua
